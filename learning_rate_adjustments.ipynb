```python
import torch
from torch import optim
from torch.optim import lr_scheduler
from torch import nn
model = nn.Linear(10,1)
optimizer = optim.SGD(model.parameters(), lr=0.0009, momentum=0.9)
scheduler0 = torch.optim.lr_scheduler.ConstantLR(optimizer, factor = 0.0009, total_iters = 2)
scheduler1 = torch.optim.lr_scheduler.LinearLR(optimizer, start_factor = 0.0002, total_iters = 5)
scheduler2 = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0 = 300, eta_min = 0.0000088, last_epoch = 300)
scheduler = torch.optim.lr_scheduler.SequentialLR(optimizer, schedulers=[scheduler0,scheduler1, scheduler2], milestones= [2,7])
```

```python
lr = []
for i in range(300):
    optimizer.step()
    scheduler.step()
    lr.append([scheduler.get_last_lr(), i])
```
